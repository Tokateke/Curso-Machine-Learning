# FUNCIONES DE DISTRIBUCIÓN DE PROBABILIDADES

## Distribución Uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1 # Origen
b = 100 # Final
n = 50000 # numero de muestras
data = np.random.uniform(a,b,n)
```

```python
%matplotlib inline 
# Es para almacenar el plot en el Notebook

plt.hist(data)
```

## Distribución Normal

```python
import numpy as np
```

```python
data2 = np.random.randn(1000000)
```

```python
x = range(1,1000001)
plt.plot(x, data2)
```

```python
plt.hist(data2)
```

```python
# Se suele hacer un plot de los datos ordenados.
# Esto nos da la distribución acumulada. 
# En el caso de una Gaussiana tiene esta forma.
# Se puede usar esto para contrastar que realmente es una Gaussiana.

plt.plot(x, sorted(data2))
```

```python
mi_media = 5.5 
desviacion_tipica = 2.5 
normal_estandar = np.random.randn(100000) 
data3 = mi_media + desviacion_tipica*normal_estandar

# Z = (X-mu)/sd =>N(0,1) 

# mu = media
# sd = desviación tipica
# Z = normal_estandar
```

```python
plt.hist(data3)
```
